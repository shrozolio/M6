```python
symbols = assets.symbol.to_list()
symbols
```

```python
END = datetime.today()
START = datetime(2022,1,1)
```

```python
datAdjClose = dat['Adj Close'].dropna()
```

```python
rets = np.log(datAdjClose.iloc[-1]/datAdjClose.iloc[0])
rets
```

```python
#index based on original order of assets just in case ID is not enough
df = pd.DataFrame(rets[symbols],columns=['MonthlyReturn'])
df
```

```python
def invest_decision(x):
    clipped_x = round(x.clip(-1,1),2)
    sumw = np.sum(abs(clipped_x))
    dec = round(clipped_x/sumw,2)
    dec[-1] = round(1 - np.sum(abs(dec[:-1])),2)
    return dec
```

```python
decision = invest_decision(df.MonthlyReturn)
```

```python
df['Decision'] = decision
```

```python
sum(abs(decision))
```

```python
df
```

```python
ranked = pd.qcut(df.MonthlyReturn,5,labels=np.arange(1,6))
ranked
```

```python
df = df.merge(pd.get_dummies(ranked, prefix="Rank", prefix_sep=""),left_index=True, right_index=True)
df
```

```python
df.index.rename("ID",inplace=True)
```

```python
new_cols = df.columns[2:].append(df.columns[[1]])
df = df[new_cols].reset_index()
df
```
