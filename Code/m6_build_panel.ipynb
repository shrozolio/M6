```python
dat = pd.read_csv(dat_fname, header=[0,1], index_col=0)
dat.index = pd.to_datetime(dat.index, format = "%Y-%m-%d")
dat.columns.names = ['Features','Stocks']
dat.index.name = 'Date'
dat.head()
```

```python
#list of trading days periods to calculate historical returns for every security
ret_lags = [20,40]
```

```python
def get_ret_cols(df, col, period = 1):
    """assumes input df with 2 levels of multiindex, calculates continuous return df of specified col"""
    ret_df = df.loc[:,(col, slice(None))].apply(np.log).diff(period)
    ret_df.columns = pd.MultiIndex.from_tuples([('Ret' + str(period), j) for (i,j) in ret_df.columns],
                                               names = ret_df.columns.names)
    return ret_df
```

```python
def get_ret_history_df(df, col, ret_lags):
    """returns dataframe with specified col of df return calc 
    uses ret_lags for list of lags to use """
    ret_lst = []
    for i in ret_lags:
        d_temp = get_ret_cols(df,col,i)
        ret_lst.append(d_temp)  
    hist_features = pd.concat(ret_lst, axis = 1)
    return hist_features
```

```python
panel = get_ret_history_df(dat,"Adj Close",ret_lags)
panel.tail()
```
