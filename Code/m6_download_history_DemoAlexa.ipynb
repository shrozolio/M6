<h3>Request Data - Output size = 'full'</h3>

```python
from alpha_vantage.timeseries import TimeSeries
import time # time delay

# Create an API object
api_key = 'UNO4CZQHSBZSN71N'
ts = TimeSeries(key=api_key, output_format = "pandas")
print(type(ts))

#Get daily OHLC data for symbols
allSym = []
for symbol in symbols:
    data_daily, meta_data = ts.get_daily(symbol, outputsize ='full')
    allSym.append(data_daily)
    time.sleep(60)

final_data = pd.concat(allSym, keys = symbols, axis = 1)
    
final_data
```

<h2>3 - A Simple Way to Download Financial Data from Investing.com in Python</h2>

<a href = "https://medium.com/geekculture/a-simple-way-to-download-financial-data-from-investing-com-in-python-8262271c804f"><b>Downloading data using investpy</b></a>

<h3>Installation</h3>

```python
# !pip install investpy
```

<h3>Import</h3>

```python
import investpy as inv
```

<h3>Request Data</h3>

```python
allTICKER_Investing = []

for symbol in symbols:
    try:
        df = inv.get_stock_historical_data(stock=symbol,
                                   country="United States",
                                   from_date="01/01/2005",
                                   to_date="05/03/2022")
        allTICKER_Investing.append(df)
        time.sleep(60)
        
    except:
        print("There is no such TICKERT" + symbol)

final_data_Investing = pd.concat(allTICKER_Investing, keys = symbols, axis = 1)
final_data_Investing
```

<h3>Save data</h3>

```python
save_fname_Investing = data_out + "Investing_API_" + pd.to_datetime('today').strftime('%Y_%m_%d') + ".csv"
save_fname_Investing
final_data_Investing.to_csv(save_fname_Investing, index=True)
```
