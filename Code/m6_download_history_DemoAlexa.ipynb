<h3>Get daily OHLC data for symbols</h3>

<a href = "https://www.alpharithms.com/python-financial-data-491110/"><b>Unofficial Python alpha_vantage API</b></a>

```python
save_fname = data_out + "alpha_vantage_API" + pd.to_datetime('today').strftime('%Y_%m_%d') + ".csv"
save_fname
```

```python
from alpha_vantage.timeseries import TimeSeries

import time # time delay

# Create an API object
ts = TimeSeries(key='UNO4CZQHSBZSN71N')
print(type(ts))


#Get daily OHLC data for symbols
allSym = {}
for symbol in symbols:
    data, meta_data = ts.get_daily(symbol)
    allSym[symbol] = pd.DataFrame.from_dict(data, orient="index").reset_index()
    time.sleep(60)
```

```python
df = (pd.DataFrame.from_dict({(i,j): allSym[i][j] 
                           for i in allSym.keys() 
                           for j in allSym[i].keys()},
                       orient='index')).T

for symbol in symbols:
    del df[( symbol,'index')]
```

```python
df.head()
```

<h3>Save df to csv</h3>

```python
df.to_csv(save_fname, index=True)
```
