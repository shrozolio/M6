```python
import numpy as np
import pandas as pd
```

```python
pd.set_option("display.max_columns", 500)
pd.set_option("display.max_rows", 500)
```

```python
data_folder  = "../DataWork/"
dat_fname    = data_folder + "yahoo_download_2022_02_19.csv" #stocks price info
out_fname    = data_folder + "yahoo_download_fixed.csv"      #fixed file to save
```

```python
dat = pd.read_csv(dat_fname, header=[0,1], index_col=0)
dat.index = pd.to_datetime(dat.index, format = "%Y-%m-%d")
dat.columns.names = ['Features','Stocks']
dat.index.name = 'Date'
dat.head()
```

```python
dat.shape
```

```python
#dat.columns.get_level_values("Features").unique()
dat.columns.levels[0]
```

```python
#IEFM has from June 2021 to Feb 2022 prices divided by 100
dat.xs("IEFM.L",level=1,axis=1).loc[dat.xs("IEFM.L",level=1,axis=1)['Adj Close'] < 100]
```

```python
dates_to_fix = dat.xs("IEFM.L",level=1,axis=1)['Adj Close'] < 100
dat.loc[dates_to_fix,(slice('Adj Close','Close'),'IEFM.L')] = \
    dat.loc[dates_to_fix,(slice('Adj Close','Close'),'IEFM.L')] * 100
dat.loc[dates_to_fix,(slice('High','Open'),'IEFM.L')] = \
    dat.loc[dates_to_fix,(slice('High','Open'),'IEFM.L')] * 100
```

```python
dat.to_csv(out_fname, index=True)
```
